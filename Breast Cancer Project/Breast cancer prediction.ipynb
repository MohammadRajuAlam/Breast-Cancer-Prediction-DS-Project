```python
# split (breake) the dataset into two parts Dependend (X) and Independend (Y) dataset
X=df.iloc[:,2:31].values # Here Dataframe conveting in list
Y=df.iloc[:,1].values    # Here Dataframe conveting in list
```

```python
# Here spliting (break) the data into two parts 80% training dataset and 20% testing dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
# feature scalinf
from sklearn.preprocessing import StandardScaler
X_train=StandardScaler().fit_transform(X_train)
X_test=StandardScaler().fit_transform(X_test)
```

```python
# Models / Algorithms

def Models(X_train,Y_train):
    
    # Here we will be use 3 algorithms which algorithms's accuracy will be good
    
    # 1) using logistic regression algorithms for find accurary result from dataset
    
    
    from sklearn.linear_model import LogisticRegression
    
    log=LogisticRegression(random_state=0)
    
    log.fit(X_train,Y_train)
    
    
    # 2) using Dession Tree algorithms for find accurary result from dataset
    
    from sklearn.tree import DecisionTreeClassifier
    
    tree=DecisionTreeClassifier(random_state=0,criterion='entropy')
    
    tree.fit(X_train,Y_train)
    
    
    # 3) using Random Forest algorithms for find accurary result from dataset
    
    from sklearn.ensemble import RandomForestClassifier
    
    forest=RandomForestClassifier(random_state=0,criterion='entropy',n_estimators=10)
    
    forest.fit(X_train, Y_train)
    
    
    print('[1] Logistic Regression accuracy:', log.score(X_train, Y_train))
    
    print('[2] Decision Tree accuracy:', tree.score(X_train, Y_train))
    
    print('[3] Random Forest accuracy:', forest.score(X_train, Y_train))
    
    return log, tree, forest # Here Returning log, tree, forest

# Here calling function model and will be display accuracy Results

model=Models(X_train,Y_train)
    
```

```python
# Testing the Model / Algorithms 

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

for i in range(len(model)):
    
    print('Model',i)
    
    print(classification_report(Y_test,model[i].predict(X_test)))
    
    print('Accuracy',accuracy_score(Y_test,model[i].predict(X_test)))
          
```
